```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
```

```python
classifier("We are happy to show you the hugging face transformers library")
```

```python
classifier("The pizza is not that great, but the crust is amazing")
```

```python
results = classifier(["We are happy to show you the transformer library", 
                      "We hope you dont hate it"])
for result in results:
  print(f"label:{result['label']}, with score: {round(result['score'], 4)}")
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
```

```python
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
# model only exists in pytorch, need to import it to tf by using 'from_pt' tag
model = TFAutoModelForSequenceClassification.from_pretrained(model_name, from_pt=True)
tokenizer = AutoTokenizer.from_pretrained(model_name)
classifier = pipeline('sentiment-analysis', model = model, tokenizer = tokenizer)
```

```python
classifier("I am a good boy")
```
