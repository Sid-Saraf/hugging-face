```python
from transformers import TFGPT2LMHeadModel, GPT2Tokenizer
```

```python
# use gpt2-large for more sophisticated language generation
tokenizer = GPT2Tokenizer.from_pretrained("gpt2-large")
model = TFGPT2LMHeadModel.from_pretrained('gpt2-large', pad_token_id=tokenizer.eos_token_id)
```

```python
sentence = ["My New M1 MacBook Air", "i am coding"]
input_ids = tokenizer.encode(sentence, return_tensors='tf')
```

```python
tokenizer.decode(input_ids[0])
```

```python
output = model.generate(input_ids, max_length = 300, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)
```

```python
output
```

```python
tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
text = tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
with open('blogpost.text', 'w') as f:
  f.write(text)
```
